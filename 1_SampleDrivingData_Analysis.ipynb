#      Analyze the Steering Angle from Generate Data
       
       Perform Exploratory Analysis on the sample training dataset for trends and patterns.
       Explore training data generated from simulator.
       
       
       # Observations : 
           Steering Angle Distribution from Udacity Data is normal distributed.
           Training data from simulator is skewed with straight and left turns.
           Make Sure to avoid skewness of the data for final Model.
           

## Load Libraries

```python

# Load libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

## Load Data

```python

# Import Driving Log
drv_log_udacity = pd.read_csv("./data/driving_log_udacity.csv",sep=',')

drv_log_student = pd.read_csv("./SimulatorTrainingData/fastest_training_6/driving_log.csv",sep=',',
                              names=['center','left','right','steering','throttle','brake','speed'])
```

## Data Summary

```python
print("============ UDACITY DATASET =======================")
print(drv_log_udacity.columns)
print("Udacity Dataset",np.shape(drv_log_udacity))
print("")


print("============ STUDENT DATASET =======================")
print(drv_log_student.columns)
print("Udacity Dataset",np.shape(drv_log_student))
```

# EXPLORATORY ANALYSIS

```python
# ======== Steering Angle Distribution  ( Univariate Analysis) ===========

canvas = plt.figure(figsize=(10,10))


# --- Udacity DataSet ---
udacity_true = canvas.add_subplot(4,2,1)
fig = udacity_true.hist(drv_log_udacity['steering'],bins=500)
udacity_true.set_title("Steering Angle Distribution - Udacity DataSet")
udacity_true.set_xlabel('Steering Angle in Degrees')
udacity_true.set_ylabel("No Of Sample")

udacity_scaled = canvas.add_subplot(4,2,3)
fig = udacity_scaled.hist(drv_log_udacity['steering'],bins=500)
udacity_scaled.set_xlabel("Steering Angle in Degrees")
udacity_scaled.set_ylabel("No Of Sample - LogScale")
udacity_scaled.set_yscale('log')

# --- Student Dataste ---
student_true = canvas.add_subplot(4,2,2)
fig = student_true.hist(drv_log_student['steering'],bins=500)
student_true.set_title("Steering Angle Distribution - Student DataSet")
student_true.set_xlabel('Steering Angle in Degrees')
student_true.set_ylabel("No Of Sample")

student_scaled = canvas.add_subplot(4,2,4)
fig = student_scaled.hist(drv_log_student['steering'],bins=500)
student_scaled.set_xlabel("Steering Angle in Degrees")
student_scaled.set_ylabel("No Of Sample - LogScale")
student_scaled.set_yscale('log')


# ============= Bivariate Analysis =============================


# ===== Steering Angle as Series =====

# --- Udacity Dataset ---
udacity_steerSeries = canvas.add_subplot(4,2,5)
fig = udacity_steerSeries.plot(drv_log_udacity['steering'], label='Steering Angle Series')
udacity_steerSeries.set_ylabel('Steering Angle')
udacity_steerSeries.set_xlabel('Samples')
plt.legend()


# --- Student Dataset ---
student_steerSeries = canvas.add_subplot(4,2,6)
fig = student_steerSeries.plot(drv_log_student['steering'], label = "Steering Angle Series")
student_steerSeries.set_ylabel('Steering Angle')
student_steerSeries.set_xlabel('Samples')
plt.legend()


# =====  Sterring Angle Vs Speed =====

# --- Udacity Dataset ---
udacity_speedvsSteer = canvas.add_subplot(4,2,7)
fig = udacity_speedvsSteer.scatter(drv_log_udacity['speed'],drv_log_udacity['steering'], label='SteeringAngle Vs Speed')
udacity_speedvsSteer.set_xlabel('Speed')
udacity_speedvsSteer.set_ylabel('Steering Angle')
plt.legend()

# --- Student Dataset ---
udacity_speedvsSteer = canvas.add_subplot(4,2,8)
fig = udacity_speedvsSteer.scatter(drv_log_student['speed'],drv_log_student['steering'], label='SteeringAngle Vs Speed')
udacity_speedvsSteer.set_xlabel('Speed')
udacity_speedvsSteer.set_ylabel('Steering Angle')
plt.legend()

# ===== Adjust  Subplot Margins =====
plt.tight_layout()
```
